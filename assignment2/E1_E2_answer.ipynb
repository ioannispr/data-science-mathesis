### Ε1: Εισόδημα και Αποτελέσματα των Τεστ


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
# φόρτωση δεδεομένων για την απόδοση των φοιτητών
country_scores=pd.read_excel('patel-sandefur-human-capital-final-results.xlsx', sheet_name='Country Scores')
country_scores
```

```python
country_scores.info()
```

```python
# φόρτωση δεδομένων για τα εισοδήματα
capital_df=pd.read_csv('WDI_data.csv')
capital_df
```

```python
capital_df.info()
```

```python
capital_df.columns
```

```python
capital_df['Series Name'].unique() # θα χρειαστούμε μ΄όνο την ετικέτα 'GDP per capita, PPP (constant 2011 international $)'
```

```python
capital_df=capital_df[capital_df['Series Name']=='GDP per capita, PPP (constant 2011 international $)']
capital_df # παρατηρούμε οτι έχουμε σαθρές τιμές
```

```python
# Θα αφαιρ΄έσουμε τις εγγραφές που έχουν null ή σαθρές (άκυρες) τιμές. Ας ελέγξουμε τι έχουμε...
```

```python
capital_df.info() # Δεν έχουμε καμία null τιμή αλλά έχουμε σαθρές τιμές
```

```python
capital_df=capital_df[capital_df['2015 [YR2015]'].str.contains(r'[a-zA-Z0-9]')] # αφαιρώ όλες τις σαθρές τιμές
capital_df
```

```python
# θα αφαιρέσουμε και όλες τις σαθρές ή null τιμές κι απο το dataframe country_scores
```

```python
country_scores.dropna(axis=0, how='any', inplace=True)
country_scores.reset_index(drop=True, inplace=True)
```

```python
# Στις παλινδρομήσεις που θα κάνουμε δεν θέλουμε να έχουμε null τιμές , διότι θα επηρεάσουν τα αποτελέσματα. 
# Επίσης παρατηρούμε ότι το dataframe 'country_scores' είναι σχετικά μικρό δείγμα.
# Γενικά στις παλινδρομήσεις θέλω όσο το δυνατόν μεγαλύτερα δείγματα.
```

```python
country_scores.shape
```

```python
capital_df.shape
```

```python
# επίσης θα φέρουμε τα δεδομένα σε ένα dataframe με inner join ως προς το "Country Code"
```

```python
len(capital_df['Country Code'].unique())==capital_df.shape[0]
```

```python
len(country_scores['Country Code'].unique())==country_scores.shape[0]
```

```python
df_for_reg=pd.merge(capital_df,country_scores,on='Country Code')
df_for_reg.info()
```

```python
df_for_reg.head()
```

```python
df_for_reg['2015 [YR2015]']= pd.to_numeric(df_for_reg['2015 [YR2015]']) # κάνουμε αριθμό το πεδίο με το εισόδημα
```

```python
df_for_reg['2015 [YR2015]'].head()
```

```python
# Μετανομασία κολωνών για διευκόλυνση
df_for_reg.rename(columns={'2015 [YR2015]':'GDP_per_Capita_in_2015',
                           'Median Math Score (TIMSS Scale, 4th Grade)':'TIMSS',
                          'Median Reading Score (PIRLS Scale, 4th Grade)':'PIRLS'},inplace=True)
```

```python
# Πρώτα θα κάνουμε την παλινδρόμηση  μεταξύ του λογαρίθμου του εισοδήματος και του σκορ TIMMS. Η εξαρτημένη μεταβλητή θα πάρει το λογάριθμο με βάση το 2 για λόγους ερμηνείας
import statsmodels.formula.api as smf
import numpy as np
df_for_reg['GDP_per_Capita_in_2015']=np.log2(df_for_reg['GDP_per_Capita_in_2015'])
log_capital_TIMSS_model=smf.ols('GDP_per_Capita_in_2015~TIMSS',data=df_for_reg)
log_capital_TIMSS_reg = log_capital_TIMSS_model.fit()
print('Οι συντελεστές της παλινδρόμησης είναι:','\n',log_capital_TIMSS_reg.params)
print('Το R^2 είναι',log_capital_TIMSS_reg.rsquared)
log_capital_TIMSS_reg.summary()
```

---
* Ερμηνεία
---

```python
# Από τα αποτελέσματα της παλινδρόμησης , βλέπουμε το R^2 είναι κοντά στο 40% , πράγμα που δείχνει ΄ότι δεν είναι και τόσο καλή η προσαρμογ΄΄η των δεδομένων μας στην 
# γραμμή παλινδρόμησης και ίσως εξήγειτε απο τα υπόλοιπα το άλλο ποσοστό.
# Ωστόσο, βλέπουμε ότι υπάρχει μια γραμμική σχέση μεταξύ εξαρτημένης και ανεξάρτητης μεταβλητής, διότι το p value του συντελεστή της εξαρτημένης
# μεταβλητής είναι πολύ κοντα στο μηδέν , πράγμα που σημαίνει οτι απορρίπτουμε την μηδενική υπόθεση για κάθε επίπεδο σημαντκότητας α.
```

```python
# Μετά θα κάνουμε την παλινδρόμηση  μεταξύ του λογαρίθμου του εισοδήματος και του σκορ PIRLS. Η εξαρτημένη μεταβλητή θα πάρει το λογάριθμο με βάση το 2 για λόγους ερμηνείας
# df_for_reg['GDP_per_Capita_in_2015']=np.log2(df_for_reg['GDP_per_Capita_in_2015'])
log_capital_PIRLS_model=smf.ols('GDP_per_Capita_in_2015~PIRLS',data=df_for_reg)
log_capital_PIRLS_reg = log_capital_PIRLS_model.fit()
print('Οι συντελεστές της παλινδρόμησης είναι:','\n',log_capital_PIRLS_reg.params)
print('Το R^2 είναι',log_capital_PIRLS_reg.rsquared)
log_capital_PIRLS_reg.summary()
```

---
* Ερμηνεία
---

```python
# Παρόμοια ερμνηνεία με την προηγούμενη , με μία σημαντική διαφορά , ότι εδώ το R^2 είναι σε καλύτερα επίπεδα
```

```python
# Γραφήματα
```

```python
fig, axes = plt.subplots(2,1, figsize=(8,16))

axes[0]=sns.scatterplot(x='GDP_per_Capita_in_2015', y='TIMSS', data=df_for_reg, ax=axes[0], color='k')
axes[0].set_xlabel('Log GDP_per_Capita_in_2015')

axes[1]=sns.scatterplot(x='GDP_per_Capita_in_2015', y='PIRLS', data=df_for_reg, ax=axes[1],color='k')
axes[1].set_xlabel('Log GDP_per_Capita_in_2015')


axes[0].grid()
axes[0].set_facecolor('azure')

axes[1].grid()
axes[1].set_facecolor('azure')


plt.show()
```

### E2: Χρόνια Εκπαίδευσης και Αποτελέσματα στα Τεστ

```python
# φόρτωση του αρχείου 
years_school=pd.read_stata('BL2013_MF1599_v2.2.dta')
years_school.info()
```

```python
# παίρνουμε μ΄όνο τις τιμές για το 2010
years_school_2010=years_school[years_school.year==2010]
years_school_2010.head()
```

```python
years_school_2010=years_school_2010.rename(columns={'yr_sch':'avg_years','country':'Country Name'})
years_school_2010.head()
```

```python
# παίρνουμε τα υπόλοιπα απο τις δύο παλινδρομήσεις που κάναμε στο ερώτημα Ε1
resid_TIMSS=log_capital_TIMSS_reg.resid
resid_PIRLS=log_capital_PIRLS_reg.resid
```

```python
capital_df=capital_df.rename(columns={'2015 [YR2015]':'GDP_per_Capita_in_2015'})
capital_df
```

```python
capital_df['GDP_per_Capita_in_2015']= pd.to_numeric(capital_df['GDP_per_Capita_in_2015']) # κάνουμε αριθμό το πεδίο με το εισόδημα
```

```python
# θα φτιάξουμε και πάλι ενα dataframe μεταξύ 'capital_df' και 'years_school_2010' για την μεταξύ τους παλινδρόμηση
df_capital_years=pd.merge(capital_df, years_school_2010, on='Country Name')
df_capital_years.info()
```

```python
df_capital_years.head(3)
```

```python
# κάνουμε την παλινδρόμηση μεταξύ 'GDP_per_Capita_in_2015~avg_years'
df_capital_years['GDP_per_Capita_in_2015']=np.log2(df_capital_years['GDP_per_Capita_in_2015'])
log_capital_years_model=smf.ols('GDP_per_Capita_in_2015~avg_years',data=df_capital_years)
log_capital_years_reg = log_capital_years_model.fit()
print('Οι συντελεστές της παλινδρόμησης είναι:','\n',log_capital_years_reg.params)
print('Το R^2 είναι',log_capital_years_reg.rsquared)
log_capital_years_reg.summary()
```

---
* Ερμηνεία
---

```python
# Όπως βλέπουμε και απο το R^2 η γραμμή παλινδρόμησης προσαρμόζεται αρκετά κάλα στα δεδομένα μας , πράγμα που σημαίνει οτι είναι ένα μοντελό , το  οποίο μπορεί
# να χρησιμεύσει και για πρόβλεψη. Ένα καλό R^2 θεωρείται κόντα ή πάνω απο το 60%.
# Τέλος έχουμε γραμμική σχέση μεταξύ των μεταβλητών επειδή απορρίπτουμε την μηδενική υπόθεση  για την ανεξάρτητη μεταβλητή.
```

```python
# παίρνουμε τα υπόλοιπα της παραπάνω παλινδρόμησης 
resid_capital_years=log_capital_years_reg.resid
```

```python
# στην συνέχεια θα κάνουμε τις παλινδρομήσεις. Πρώτα θα τα βάλουμε όλα ένα dataframe, το οποίο θα έχει και null τιμές που θα τις εξαλείψουμε
df_resid_TIMSS_PIRLS=pd.DataFrame({'resid_TMISS':resid_TIMSS,'resid_PIRLS':resid_PIRLS})
df_resid_capital_years=pd.DataFrame({'resid_capital_years':resid_capital_years})
df_all_resid=pd.concat([df_resid_TIMSS_PIRLS, df_resid_capital_years], axis=1)
df_all_resid.dropna(axis=0,how='any', inplace=True)
df_all_resid.head()
```

```python
# 'resid_capital_years~resid_TIMSS'
log_resid_capital_resid_TIMSS_model=smf.ols('resid_TMISS~resid_capital_years',data=df_all_resid)
log_resid_capital_resid_TIMSS_reg = log_resid_capital_resid_TIMSS_model.fit()
print('Οι συντελεστές της παλινδρόμησης είναι:','\n',log_resid_capital_resid_TIMSS_reg.params)
print('Το R^2 είναι',log_resid_capital_resid_TIMSS_reg.rsquared)
log_resid_capital_resid_TIMSS_reg.summary()
```

---
* Ερμηνεία
---

```python
# Τόσο  το R^2 όσο και το p-value της συγκεκριμένης παλινδρόμησης δεν είναι ενθαρρυντικά για το μοντέλο μας, δίοτι δεν έχουμε καλή προσαρμογή των δεδομένων στην γραμμή
# παλινδρόμησης αλλά και απορρίπτουμε την μηδενική υπόθεση της ανεξάρτητης μεταβλτ΄ής μόνο σε επίπεδο σημαντικότητας α=10%.
```

```python
# 'resid_capital_years~resid_TIMSS'
log_resid_capital_resid_PIRLS_model=smf.ols('resid_PIRLS~resid_capital_years',data=df_all_resid)
log_resid_capital_resid_PIRLS_reg = log_resid_capital_resid_PIRLS_model.fit()
print('Οι συντελεστές της παλινδρόμησης είναι:','\n',log_resid_capital_resid_PIRLS_reg.params)
print('Το R^2 είναι',log_resid_capital_resid_PIRLS_reg.rsquared)
log_resid_capital_resid_PIRLS_reg.summary()
```
